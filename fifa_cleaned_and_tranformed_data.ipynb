```python
# def function to clear and transform the values.
def clean_value(value):
    value = value.replace('€', '')
    if 'M' in value:
        value = float(value.replace('M', '')) * 1000000
    elif 'K' in value:
        value = float(value.replace('K', '')) * 1000
    return int(value)
```

```python
# Applying the function to the columns
df_project1['Value'] = df_project1['Value'].apply(clean_value)
df_project1['Wage'] = df_project1['Wage'].apply(clean_value)
```

# 4. Removing characters from Hits column

```python
df_project1['Hits'].head()
```

```python
#def function to remove newlines from Hits column
def clean_and_convert_to_int(hits):
    hits = hits.replace('\n', '').strip()
    multiplier = 1
    if 'K' in hits:
        multiplier = 1000
        hits = hits.replace('K', '')
    return int(float(hits) * multiplier)
```

```python
df_project1['Hits'] = df_project1['Hits'].apply(clean_and_convert_to_int)
```

```python
df_project1['Hits'].head()
```

# 5. separating the column Joined into Year, Month and Day columns.

```python
df_project1['Joined'].head(10)
```

```python
# Converting the 'Joined' column to datetime type
df_project1['Joined'] = pd.to_datetime(df_project1['Joined'])
```

# 6. Separating Team & Contract column

```python
df_project1['Team & Contract'].head(10)
```

```python
def clean_and_split_team_contract(data):
    data = data.strip()  # Remove leading and trailing whitespaces
    team, contract = data.split('\n')[-2:]  # Split by '\n' and take last two elements
    return pd.Series([team, contract], index=['Team', 'Contract'])
```

```python
df_project1[['Team', 'Contract']] = df_project1['Team & Contract'].apply(clean_and_split_team_contract)
```

```python
df_project1 = df_project1.drop('Team & Contract', axis = 1)
```
